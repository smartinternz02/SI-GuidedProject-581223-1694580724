```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
columnsobject=['Attrition', 'BusinessTravel',  'Department', 'EducationField',
       'Gender',  
      'JobRole', 'MaritalStatus',
        'Over18',
       'OverTime']
for i in range(len(columnsobject)):
    data[columnsobject[i]]=le.fit_transform(data[columnsobject[i]])
data.head()
```

```python
data.info()
```

```python
from sklearn.model_selection import train_test_split
x=data.drop('Attrition',axis=1)
y=data.Attrition
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lrr=LogisticRegression()
```

```python
lrr.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,roc_auc_score,f1_score
```

```python
y_pred=lrr.predict(x_test)
```

```python
print("Training Score: ",lrr.score(x_train,y_train))
print("Testing Score: ",lrr.score(x_test,y_test))
```

```python
print("Accuracy: ",accuracy_score(y_test,y_pred))
print("Confusion Matrix: ",confusion_matrix(y_test,y_pred))
print("Precision: ",precision_score(y_test,y_pred))
print("Recall: ",recall_score(y_test,y_pred))
print("AOC RUC: ",roc_auc_score(y_test,y_pred))
print("F1: ",f1_score(y_test,y_pred))
```

# Grid SearchCV For Logistic regression

```python
from sklearn.model_selection import GridSearchCV
hyperparameters = {
    'penalty': ['l1', 'l2'],
    'C': [0.01, 0.1, 1, 10,34,56,78,12,36,1000,20, 100,50,75]
}
```

```python
grid_search = GridSearchCV(lrr, hyperparameters, cv=10)
```

```python
grid_search.fit(x1_train,y1_train)
```

```python
grid_search.best_params_
```

```python
grid_search.best_score_
```

```python
lr1=LogisticRegression(C=grid_search.best_params_.get('C'),penalty=grid_search.best_params_.get('penalty'))
```

```python
lr1.fit(x1_train,y1_train)
```

```python
yg_pred=lr1.predict(x1_test)
```

```python
print("Accuracy: ",accuracy_score(y1_test,yg_pred))
print("Confusion Matrix: ",confusion_matrix(y1_test,yg_pred))
print("Precision: ",precision_score(y1_test,yg_pred))
print("Recall: ",recall_score(y1_test,yg_pred))
print("AOC RUC: ",roc_auc_score(y1_test,yg_pred))
print("F1: ",f1_score(y1_test,yg_pred))
```
